```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_films_set_in_the_future').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```python
links = My_table.findAll('a')
links
```

```python
Titles = []
for link in links:
    Titles.append(link.get('title'))
    
print(Titles)
```

```python
import pandas as pd
df = pd.DataFrame()
df['Title'] = Titles

df
```
