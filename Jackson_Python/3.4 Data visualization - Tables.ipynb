```python
#for table 1
UL_pivoted_city_share = UL_city_summary_table_AR_share.pivot(index ='tweet_date', columns ='city', values =['lang_share'])
UL_pivoted_city_share = UL_pivoted_city_share['lang_share']
UL_pivoted_city_share = UL_pivoted_city_share[['gaziantep', 'budapest', 'munich', 'berlin','paris']]
UL_pivoted_city_share = UL_pivoted_city_share.fillna(0)

#for table 2
UL_pivoted_date_share = UL_city_summary_table_AR_share.pivot(index ='city', columns ='tweet_date', values =['lang_share'])
UL_pivoted_date_share = UL_pivoted_date_share.loc[['gaziantep', 'budapest', 'munich', 'berlin','paris']]
```

```python
#I. Using pandas Styler function to visualize table data
final_table_1 = UL_pivoted_city_share.style\
                  .apply(lambda x: ["background: ghostwhite" if v == 0 else "" for v in x], axis = 0)\
                  .format({"gaziantep": "{0:.2%}", "budapest": "{0:.2%}", "munich": "{:.2%}", "berlin": "{:.2%}", "paris": "{:.2%}"})\
                  .bar(subset=["gaziantep",], color='lavender')\
                  .bar(subset=["budapest",], color='royalblue')\
                  .bar(subset=["munich",], color='lightsteelblue')\
                  .bar(subset=["berlin",], color='mediumpurple')\
                  .bar(subset=["paris",], color='thistle')\
                  .set_properties(**{'background-color': 'whitesmoke'})
```

```python
final_table_1_html = (final_table_1.render())
```

```python
final_table_1_html_text = open("final_table_1.html", "a")
final_table_1_html_text.write(final_table_1_html)
final_table_1_html_text.close()
```

```python
# Table with conditional formatting
final_table_2 = UL_pivoted_city_share.style\
                  .format({"gaziantep": "{0:.2%}", "budapest": "{0:.2%}", "munich": "{:.2%}", "berlin": "{:.2%}", "paris": "{:.2%}"})\
                  .set_properties(**{'background-color': 'whitesmoke'})\
                  .apply(lambda x: ["background: ghostwhite" if v == 0 else "" for v in x], axis = 0)\
                  .apply(lambda x: ["background: plum" if v <= 0.0005 else "" for v in x], axis = 1)\
                  .apply(lambda x: ["background: thistle" if v > 0.0005 and v <= 0.001 else "" for v in x], axis = 1)\
                  .apply(lambda x: ["background: mediumorchid" if v > 0.001 and v <= 0.005 else "" for v in x], axis = 1)\
                  .apply(lambda x: ["background: mediumvioletred" if v > 0.005 and v <= 0.01 else "" for v in x], axis = 1)\
                  .apply(lambda x: ["background: darkviolet" if v > 0.01 and v <= 0.03 else "" for v in x], axis = 1)\
                  .apply(lambda x: ["background: darkorchid" if v > 0.03 else "" for v in x], axis = 1)
```

```python
final_table_2_html = (final_table_2.render())
```

```python
final_table_2_html_text = open("final_table_2.html", "a")
final_table_2_html_text.write(final_table_2_html)
final_table_2_html_text.close()
```

### 3.4.2 Using Tweet Level Data to create time-series tables

```python
# Grouoping df_final on city, lang and tweet, and count tweets per language
TL_city_summary_table_AR = df_final[df_final.lang == 'ar']\
                                             .groupby(["tweet_date","city","lang"])["lang"].count()\
                                             .reset_index(name="day_total_count")

TL_city_summary_table_TR = df_final[df_final.lang == 'tr']\
                                             .groupby(["tweet_date","city","lang"])["lang"].count()\
                                             .reset_index(name="day_total_count")

TL_city_summary_table_FR = df_final[df_final.lang == 'fr']\
                                             .groupby(["tweet_date","city","lang"])["lang"].count()\
                                             .reset_index(name="day_total_count")

TL_city_summary_table_EN = df_final[df_final.lang == 'en']\
                                             .groupby(["tweet_date","city","lang"])["lang"].count()\
                                             .reset_index(name="day_total_count")
```

```python
# Pulling needed columns for final data visualization
TL_city_summary_table_AR2 = TL_city_summary_table_AR[["city","tweet_date","day_total_count"]]
TL_city_summary_table_TR2 = TL_city_summary_table_TR[["city","tweet_date","day_total_count"]]
TL_city_summary_table_FR2 = TL_city_summary_table_FR[["city","tweet_date","day_total_count"]]
TL_city_summary_table_EN2 = TL_city_summary_table_EN[["city","tweet_date","day_total_count"]]
```

```python
# Two table formats
## table 1: Rows = tweet_date, Columns = city
TL_pivoted_city_AR = TL_city_summary_table_AR2.pivot(index ='tweet_date', columns ='city', values =['day_total_count'])
TL_pivoted_city_AR = TL_pivoted_city_AR['day_total_count']
TL_pivoted_city_AR = TL_pivoted_city_AR[['gaziantep', 'budapest', 'munich', 'berlin']]


TL_pivoted_city_TR = TL_city_summary_table_TR2.pivot(index ='tweet_date', columns ='city', values =['day_total_count'])
TL_pivoted_city_TR = TL_pivoted_city_TR['day_total_count']
TL_pivoted_city_TR = TL_pivoted_city_TR[['gaziantep', 'budapest', 'munich', 'berlin']]


TL_pivoted_city_EN = TL_city_summary_table_EN2.pivot(index ='tweet_date', columns ='city', values =['day_total_count'])
TL_pivoted_city_EN = TL_pivoted_city_EN['day_total_count']
TL_pivoted_city_EN = TL_pivoted_city_EN[['gaziantep', 'budapest', 'munich', 'berlin']]


## table 2: Rows = city, Columns = tweet_date
TL_pivoted_date_AR = TL_city_summary_table_AR2.pivot(index ='city', columns ='tweet_date', values =['day_total_count'])
TL_pivoted_date_AR = TL_pivoted_date_AR.loc[['gaziantep', 'budapest', 'munich', 'berlin'], :]
```

```python
TL_pivoted_city_AR.style.format({"gaziantep": "{:}", "budapest": "{:}", "munich": "{:}", "berlin": "{:}"})\
                  .bar(subset=["gaziantep",], color='lavender')\
                  .bar(subset=["budapest",], color='royalblue')\
                  .bar(subset=["munich",], color='lightsteelblue')\
                  .bar(subset=["berlin",], color='firebrick')
```
