## 1.1 Collection - Get Old Tweets

```python
def GOTcity(start_date, stop_date, location, file_name):
    """Dates in 'yyyy-mm-dd' format; Location in decimals '52.52, 13.40', filename in 'Berlin20200410.csv' format"""
    with open(file_name, 'w', encoding="utf-8") as f:
        tweetCriteria = got.manager.TweetCriteria().setMaxTweets(9999).setSince(start_date).setUntil(stop_date).setNear(location).setWithin("35km");
        tweets = got.manager.TweetManager.getTweets(tweetCriteria);
        for tweet in tweets:
            if len(tweet.text) > 10:
                f.write(f"{tweet.id},{tweet.date},{tweet.username},{tweet.text.replace(',', '')}'\n'")
```

## 2.2 Assigning a primary language to each user 

```python
##Apraoch: How many langauges is each user tweeting in? 
df_final_user_subset = df_final.groupby(["tweet_username","lang"])["lang"]\
                      .count()\
                      .reset_index(name="user_lang_count")\
                      .sort_values(by='tweet_username', ascending=False)\
                      .dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

df_final_user_total = df_final_user_subset.groupby(["tweet_username"])["tweet_username"].count().reset_index(name="user_tweet_count").sort_values(by='user_tweet_count', ascending=False)
df_final_user_subset = df_final_user_subset.merge(df_final_user_total, left_on='tweet_username', right_on='tweet_username')
#calcualte lang share of total tweets
df_final_user_subset['user_lang_share'] = df_final_user_subset.user_lang_count/df_final_user_subset.user_tweet_count
#Rank languages by share
df_final_user_subset['user_lang_rank'] = df_final_user_subset.groupby("tweet_username")["user_lang_share"].rank("first", ascending=False)
#Keeping only Rank 1
df_final_user_subset = df_final_user_subset[(df_final_user_subset['user_lang_rank'] == 1)] # & (df_final_user_subset.tweet_username == 'kyriii')
df_user_primary_lang = df_final_user_subset.rename(columns={"lang": "primary_lang"}, errors="raise")
df_user_primary_lang = df_user_primary_lang.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
#Join primary lang
df_final_primary_lang = pd.merge(df_final.drop_duplicates(),df_user_primary_lang.drop_duplicates()[['tweet_username','primary_lang']]\
                                 , on='tweet_username'\
                                 , how='left'\
                                )
df_final_primary_lang.shape
```

```python
#limit DataFrame to only include one tweet per user
df_final_user_primary_lang = df_final_primary_lang.groupby(["tweet_username","primary_lang","city","tweet_date"])["tweet_text"]\
                      .count()\
                      .reset_index(name="user_tweet_count")\
                      .sort_values(by='city', ascending=False)\
                      .dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
df_final_user_primary_lang.shape
```

```python
# #Augmenting df_final_primary_lang
city_total = df_final_primary_lang.groupby(["city","tweet_date"], as_index = False)["tweet_id"].count().rename(columns={"tweet_id": "city_total"})
city_lang_total = df_final_primary_lang.groupby(["city","primary_lang","tweet_date"], as_index = False)["tweet_id"].count().rename(columns={"tweet_id": "city_lang_total"})
df_final_user_primary_lang = pd.merge(\
                                        df_final_user_primary_lang, city_lang_total\
                                        , on = ["city","primary_lang","tweet_date"]\
                                        , how = "left"\
                                       )
df_final_user_primary_lang = pd.merge(\
                                        df_final_user_primary_lang, city_total\
                                        , on = ["city","tweet_date"]\
                                        , how = "left"\
                                       )

df_final_user_primary_lang["lang_share"] = df_final_user_primary_lang.city_lang_total/df_final_user_primary_lang.city_total
#df_final_primary_lang_subset[df_final_primary_lang_subset["city"] == "gaziantep"].head(30)
```
