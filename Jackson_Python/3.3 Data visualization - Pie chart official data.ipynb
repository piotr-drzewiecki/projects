## 3.3 Pie chart official data

```python
#loanding pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re as re
import datetime as datetime

%matplotlib inline 

colormap=plt.cm.Paired
```

```python
#READ IN CSV HERE>>
df_2018Munichofficial = pd.read_csv('C:/Users/juans/Desktop/Python Final Project/Munich_2018_official.csv')
```

```python
df_2018Munichofficial['Explode'] = 0
df_2018Munichofficial.loc[df_2018Munichofficial.lang == 'tr', 'Explode'] = 0.15
df_2018Munichofficial.loc[df_2018Munichofficial.lang == 'ar', 'Explode'] = 0.15
df_2018Munichofficial
```

```python
plt.style.use("bmh")
plt.figure(figsize=(17,12))

slices = df_2018Munichofficial['lang_count']
labels = df_2018Munichofficial['lang']
explode = df_2018Munichofficial['Explode']
colors = [colormap(i) for i in np.linspace(0,1,12)]

plt.pie(slices, explode=explode, labels=labels, labeldistance=1.15,
        textprops=dict(size=18),  
        autopct='%1.f%%', pctdistance=1.07,
        radius=1.05, colors=colors)

#DONT FORGET TO CHANGE TITLE YEAR
plt.title("Munich - % Languages According to Nationalities 2018", weight="bold", 
          color="black", size=40)
# plt.tight_layout()
plt.show()
```
